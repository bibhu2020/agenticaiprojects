# What is MCP (Model Context Protocol)?
- A protocol to integrate with shared **tools**, resources, and prompts developed by others.
- "A USB-C port for AI applications"

![MCP](https://www.descope.com/_next/image?url=https%3A%2F%2Fimages.ctfassets.net%2Fxqb1f63q68s1%2F2x3R1j8peZzdnweb5m1RK3%2Fa8628561358334a605e7f291560fc7cc%2FMCP_learning_center_image_1-min__1_.png&w=1920&q=75)

## Core Components

**Host**: It is an LLM App like Claude or our agent architecture.

**MCP Client**: lives inside the **Host** and connects 1:1 to the **MCP Server**.

**MCP Server**: provides tools, context and prompts. It can either live within the same machine where yo have the host, or on a remote machine. MCP server might be from a 3rd party, but you can download and run from your machine.

![mcp](https://images.ctfassets.net/xqb1f63q68s1/tED7RTeh1GJfWR3x7DuZJ/7e79ee0829dcbb039ab6dc635f2dbf32/MCP_core_components-min.png)

### Examples:

- **Fetch** is an **MCP server** that searches the web via a headless browser. It opens a headless browser, uses microsoft playright to drive the browser, and collect the page content.

- You can configure a **Claude Desktop** (the **Host**) to run an **MCP Client** that then launches the Fetch **MCP server** on your computer.

- It is your **MCP Client** that typically launches **MCP Server** within the same **Host**. It's rare, people keep it on the remote server.

## Transport Mechanisms

- **Stdio (Standard IO)**: **MCP Client** spawns a process (running the **MCP server** locally) and communicates with that process via standard input/output. It is very common. 

- **SSE (Service Side Event)**: uses https connection with streaming. It is used when you have the **MCP Server** running on the Remote Server.
